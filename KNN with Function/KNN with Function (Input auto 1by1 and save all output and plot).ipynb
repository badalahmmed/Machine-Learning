```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
from sklearn.metrics import f1_score
import os
from glob import glob
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
def knn_NAHID(lnk,file_name):
  # Importing the datasets
  datasets2 = pd.read_csv(lnk)

  X2 = datasets2.iloc[:, 0:4].values
  Y2 = datasets2.iloc[:, 5].values


  # Splitting the dataset into the Training set and Test set
  X_Train2, X_Test2, Y_Train2, Y_Test2 = train_test_split(X2, Y2, test_size = 0.30, random_state = 0)


  #Create KNN Classifier
  knn = KNeighborsClassifier(n_neighbors=5)


  #Train the model using the training sets
  knn.fit(X_Train2, Y_Train2)

  #Predict the response for test dataset
  Y_Pred2 = knn.predict(X_Test2) #TEST_DATA

  # Model Accuracy, how often is the classifier correct?
  Accuracy=metrics.accuracy_score(Y_Test2, Y_Pred2) #(TRUE, PREDICTED_LABEL)
  Precision=metrics.precision_score(Y_Test2, Y_Pred2)
  Recall=metrics.recall_score(Y_Test2, Y_Pred2)
  F1_Score=metrics.f1_score(Y_Test2, Y_Pred2)
  Cohen_Kappa_Score=metrics.cohen_kappa_score(Y_Test2, Y_Pred2, labels=None, weights=None, sample_weight=None)

  return file_name, Accuracy, Precision, Recall, F1_Score, Cohen_Kappa_Score
```

```python
all_result_knn=[]
PATH = "/content/drive/My Drive/Emotiv_data/csv/"
EXT = "*.csv"
all_csv_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]

count=0
for i in all_csv_files:
  all_result_knn.append(knn_NAHID(i,Path(i).stem))
  count+=1
  print(count)

with open('k-nn_result.txt', 'w') as f:
    for item in all_result_knn:
      f.write("\n\n\n")
      for i in item:
        f.write("%s\n" % str(i))
```

```python
accuracy=[]
for i in range(len(all_result_knn)):
  accuracy.append(all_result_knn[i][1])
```
