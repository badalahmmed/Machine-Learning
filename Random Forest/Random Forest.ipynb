```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score 
```

```python
dataset = pd.read_csv('data')
```

```python
X = dataset.iloc[:, 0:63].values 
y = dataset.iloc[:, 64].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling 
sc = StandardScaler() 
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test)
'''
classifier = RandomForestClassifier(n_estimators=20, random_state=0) 
classifier.fit(X_train, y_train) 
y_pred = classifier.predict(X_test)
'''
X, y = make_classification(n_samples=7397, n_features=10, n_informative=2, n_redundant=0, random_state=0, shuffle=False)
clf = RandomForestClassifier(max_depth=100, random_state=2)
 
clf.fit(X_train, y_train)  #clf.fit(train_data, train_label)
y_pred = clf.predict(X_test) #print(clf.predict(train_label_set)
```

```python
#Use it or can use the performance evaluation part from the SVM model## 
print(confusion_matrix(y_test,y_pred)) 
print(classification_report(y_test,y_pred)) 
print(accuracy_score(y_test, y_pred))
```
